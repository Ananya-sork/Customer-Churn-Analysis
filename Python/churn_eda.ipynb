```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display all columns
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv("Telco_customer_churn_clean.csv")
```

```python
# Distribution of Churn
sns.countplot(x='ChurnLabel',data=df)

plt.title("Customer Churn Distribution")

plt.show()
```

observation: Most customers have not churned.

BI(Business Insight): Around three-quarters of customers stayed, while roughly one-quarter churned.

RA(Recommended Action): Focus retention efforts on customers most likely to leave.

```python
# Gender Distribution
sns.countplot(x='Gender',data=df)

plt.title("Gender Distribution")

plt.show()
```

observation: Male and female customers are almost equal.

```python
# Churn by Gender
sns.countplot(x='Gender', hue='ChurnLabel', data=df)

plt.title("Gender vs Churn")

plt.show()
```

observation: Does gender affect churn? No.

BI: Churn rates are similar across genders.

RA: Gender-specific marketing is unlikely to reduce churn.

```python
# Contract Type
sns.countplot(x='Contract', hue='ChurnLabel', data=df)

plt.xticks(rotation=20)

plt.show()
```

observation: Month-to-month customers churn much more.

BI: Month-to-month customers have the highest churn.

RA: Encourage annual or two-year contracts with discounts.

```python
# Payment Method
plt.figure(figsize=(10,5))

sns.countplot(x='PaymentMethod', hue='ChurnLabel', data=df)

plt.xticks(rotation=45)

plt.show()
```

observation: Electronic check users churn the most.

BI: Electronic check users churn more than other payment methods.

RA: Investigate issues with this payment experience and promote auto-pay.

```python
# Internet Service
sns.countplot(x='InternetService', hue='ChurnLabel', data=df)

plt.show()
```

observation: Fiber customers churn more.

BI: Fiber customers churn more frequently.

RA: Review pricing, service quality, and customer support for fiber plans.

```python
# Senior Citizen
sns.countplot(x='SeniorCitizen', hue='ChurnLabel', data=df)

plt.show()
```

observation: Senior citizens have relatively higher churn.

BI: Senior citizens show a relatively higher churn rate.

RA: Provide simplified support and targeted retention campaigns.

```python
# Monthly Charges Histogram
plt.figure(figsize=(8,5))

plt.hist(df['MonthlyCharges'],bins=30)

plt.xlabel("Monthly Charges")

plt.ylabel("Customers")

plt.title("Monthly Charges Distribution")

plt.show()
```

observation: Understand where most customers fall.

BI: Customers with higher monthly charges tend to churn more.

RA: Offer personalized discounts or bundled plans to high-value customers.

```python
# Boxplot
sns.boxplot(x='ChurnLabel', y='MonthlyCharges', data=df)

plt.show()
```

observation: Customers with high monthly charges churn more.

```python
# Tenure Histogram
plt.hist(df['TenureMonths'],bins=30)

plt.xlabel("Tenure in Months")

plt.ylabel("Customers")

plt.show()
```

obs: Many customers are relatively new.

BI: Newer customers churn more often than long-term customers.

RA: Improve onboarding and engage customers during the first few months.

```python
# Churn vs Tenure
sns.boxplot(x='ChurnLabel', y='TenureMonths', data=df)

plt.show()
```

obs: Customers who stay longer churn less.

```python
# Correlation Heatmap

# Only numeric columns.
numeric_df = df.select_dtypes(include=['int64','float64'])

# Then
plt.figure(figsize=(10,8))

sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')

plt.show()
```

obs: Look for strong positive or negative correlations.

```python
# Pie Chart
df['ChurnLabel'].value_counts().plot(kind='pie', autopct='%1.1f%%')

plt.ylabel("")

plt.show()
```

## Final Summary
1. Month-to-month contracts are the strongest indicator of customer churn.
2. Customers paying through electronic check are more likely to leave.
3. Higher monthly charges are associated with increased churn.
4. Customers with shorter tenure are at greater risk of churning.
5. Long-term contracts significantly improve customer retention.
6. Fiber internet customers require targeted retention strategies.
