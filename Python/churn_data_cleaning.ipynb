Checking blank rows

```python
(df['TotalCharges'] == " ").sum()
```

```python
# replacing blanks with NaN
df['TotalCharges'] = df['TotalCharges'].replace(" ", np.nan)
```

```python
# Checking again
df.isnull().sum()

# TotalCharges shows 11
```

Convert TotalCharges into Number

```python
# currently it's object, got to know from df.info()

df['TotalCharges'] = pd.to_numeric(df['TotalCharges'])

# now it's float64
```

Fill Missing Values

```python
# fill them with median.
df['TotalCharges'].fillna(df['TotalCharges'].median(), inplace=True)

# Check: df.isnull().sum() -> 0
```
